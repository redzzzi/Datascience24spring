# Variability

Standard Deviation

```python
# Make an array any_number
any_numbers = np.array([1, 2, 2, 10])
```

```python
# Step 1. The average.
mean = np.mean(any_numbers)
mean
# 3.75
```

```python
# Step 2. The deviations from average.
deviations = any_numbers - mean
calculation_steps = pd.DataFrame({
        'Value': any_numbers,
        'Deviation from Average': deviations
})
calculation_steps
```

```python
sum(deviations)
```

```python
np.mean(deviations)
# 0.0
```

```python
# Step 3. The squared deviations from average
squared_deviations = deviations ** 2
calculation_steps['Squared Deviations from Average'] = squared_deviations
calculation_steps
```

```python
# Step 4. Variance = the mean squared deviation from average
variance = np.mean(squared_deviations)
variance
# 13.1875
```

```python
# Step 5.
# Standard Deviation:    root mean squared deviation from average
# Steps of calculation:   5    4      3       2             1
sd = variance ** 0.5
sd
# 3.6314597615834874
```

```python
np.std(any_numbers)
```

# The SD and the Normal Curve

Chebyshev's Bounds: "Tail" Bounds

```python
def chebyshev(num_SDs):
    """
    Returns a LOWER BOUND on the fraction of the data
    within num_SDs of the mean.

    Put another way AT LEAST chebyshev(z) data is within mean +/- z * SDs.
    """
    return 1 - 1 / (num_SDs**2)
```

```python
# Proportion in the range "mean +/- 2 SDs"
chebyshev(2)
```

```python
# Proportion in the range "mean +/- 3.5 SDs"
chebyshev(3.5)
```

```python
# Proportion in the range "mean +/- 1.1 SDs"
chebyshev(1.1)
```

What does Chebyshev say?


```python
# Chebyshev's bound:
# This proportion should be at least
print("Chebyshev says that at least", \
      round(chebyshev(3),2), \
      "of the rows will be within 3 standard deviations of the mean")
```

```python
 def cheby_says(table, label, z):
    temp = table[label]
    mean = np.mean(temp)
    sd = np.std(temp)

    chosen = table[(mean - z*sd < table[label]) & \
            (table[label] < mean + z*sd) ]
    proportion = len(chosen) / len(table)
    percent = round(proportion * 100, 2)
    print('Average plus or minus', z, 'SDs:', \
              percent, '%', " which is > ", \
              np.round(chebyshev(z) *100, 2), "(Chebyshev)")
```

```python
cheby_says(births, 'Maternal Pregnancy Weight', 3)
```

```python
for label in births.columns:
    print()
    print(label)
    for z in np.array([2, 3, 4, 5]):
        cheby_says(births, label, z)
```

```python
def standard_units(x):
    """Convert array x to standard units."""
    return (x - np.mean(x)) / np.std(x)
```
